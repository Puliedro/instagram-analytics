- **Visualization:** Line graph showing cumulative follower growth day by day or week by week.

```python
# Import necessary libraries
import pandas as pd
import plotly.graph_objects as go
import json

# Load and read the accounts JSON data
with open('data/instagram_analytics_accounts.json') as f:
    accounts_data = json.load(f)

# Convert to pandas DataFrame
accounts_df = pd.DataFrame(accounts_data)

# Convert 'run_time' to datetime format
accounts_df['run_time'] = pd.to_datetime(accounts_df['run_time'])

# Remove duplicates for the same timestamp (optional)
accounts_df = accounts_df.drop_duplicates(subset='run_time', keep='last')

# Set the index to 'run_time' for easy time-based operations
accounts_df.set_index('run_time', inplace=True)

# Resample the data by 12-hour intervals and take the last available follower count every 12 hours
follower_growth_12h = accounts_df['follower_count'].resample('12H').last()

# Reset index to have a clean dataframe for plotting
follower_growth_12h = follower_growth_12h.reset_index()

# Create a plotly figure for follower growth (every 12 hours) without animation
fig_growth_12h = go.Figure()

# Add the follower growth line (interactive)
fig_growth_12h.add_trace(go.Scatter(x=follower_growth_12h['run_time'], 
                         y=follower_growth_12h['follower_count'],
                         mode='lines+markers',
                         name='Follower Growth (12-hour intervals)',
                         marker=dict(size=6, color='deepskyblue'),  # Updated color to fit dark theme
                         line=dict(width=2, color='deepskyblue'),  # Updated color to fit dark theme
                         hovertemplate='Date: %{x}<br>Follower Count: %{y}<extra></extra>'  # Custom hover info
                         ))

# Customize the layout for dark theme and better visualization
fig_growth_12h.update_layout(
    title='Follower Growth Over Time (Every 12 Hours)',
    title_x=0.5,
    hovermode='x',
    template='plotly_dark',  # Dark theme applied
    xaxis_title='Date',
    yaxis_title='Follower Count'
)

# Show the static plot without animation
fig_growth_12h.show()
```

```python
# Import necessary libraries
import pandas as pd
import plotly.graph_objects as go
import json

# Load and read the accounts JSON data
with open('data/instagram_analytics_accounts.json') as f:
    accounts_data = json.load(f)

# Convert to pandas DataFrame
accounts_df = pd.DataFrame(accounts_data)

# Convert 'run_time' to datetime format
accounts_df['run_time'] = pd.to_datetime(accounts_df['run_time'])

# Remove duplicates for the same timestamp (optional)
accounts_df = accounts_df.drop_duplicates(subset='run_time', keep='last')

# Set the index to 'run_time' for easy time-based operations
accounts_df.set_index('run_time', inplace=True)

# Resample the data by 12-hour intervals and take the last available follower count every 12 hours
follower_growth_12h = accounts_df['follower_count'].resample('12H').last()

# Calculate 12-hour follower change (gained/lost every 12 hours)
follower_change_12h = follower_growth_12h.diff().fillna(0)  # Difference between consecutive 12-hour intervals

# Reset index to have a clean dataframe for plotting
follower_change_12h = follower_change_12h.reset_index()

# Separate gains and losses for visualization
gains_12h = follower_change_12h[follower_change_12h['follower_count'] >= 0]
losses_12h = follower_change_12h[follower_change_12h['follower_count'] < 0]

# Create a plotly figure for follower change (gained/lost every 12 hours)
fig_change_12h = go.Figure()

# Add a bar chart for 12-hour follower gains (positive changes)
fig_change_12h.add_trace(go.Bar(x=gains_12h['run_time'], 
                            y=gains_12h['follower_count'],
                            marker_color='deepskyblue',  # Positive values (gains) in green
                            name='Followers Gained (12-hour intervals)',
                            hovertemplate='Date: %{x}<br>Followers Gained: %{y}<extra></extra>'  # Custom hover info
                            ))

# Customize layout for the change graph
fig_change_12h.update_layout(
    title='Follower Gains (Every 12 Hours)',
    title_x=0.5,
    hovermode='x',
    template='plotly_dark',  # Apply dark theme for consistency
    xaxis_title='Date',
    yaxis_title='Number of Followers',
    barmode='overlay',  # To overlay gains and losses for better comparison
)

# Show the plot
fig_change_12h.show()
```

# **Average Interactions per Post**.
- Description: Calculate the average number of likes and comments per post.
- Visualization: Numerical indicators and bar charts with average likes and comments per post.

1. Average Interactions per Post (Numerical Indicator)

```python
import pandas as pd
import plotly.graph_objects as go
import json

# Load and read the posts JSON data
with open('data/instagram_analytics_posts.json') as f:
    posts_data = json.load(f)

# Convert to pandas DataFrame
posts_df = pd.DataFrame(posts_data)

# Calculate the average likes and comments per post
average_likes = posts_df['like_count'].mean()
average_comments = posts_df['comment_count'].mean()
total_posts = len(posts_df)  # Total number of posts

# Create a plotly figure for the numerical indicators
fig_indicator = go.Figure()

# Add an indicator for the total number of posts
fig_indicator.add_trace(go.Indicator(
    mode="number",
    value=total_posts,
    title={"text": "Total Number of Posts"},
    domain={'x': [0, 0.33], 'y': [0, 1]}  # Position to the left
))

# Add an indicator for the average likes
fig_indicator.add_trace(go.Indicator(
    mode="number",
    value=average_likes,
    title={"text": "Average Likes"},
    domain={'x': [0.33, 0.66], 'y': [0, 1]}  # Position in the center
))

# Add an indicator for the average comments
fig_indicator.add_trace(go.Indicator(
    mode="number",
    value=average_comments,
    title={"text": "Average Comments"},
    domain={'x': [0.66, 1], 'y': [0, 1]}  # Position to the right
))

# Customize layout
fig_indicator.update_layout(
    title='Average Interactions per Post',
    title_x=0.5,
    template='plotly_dark'  # Apply dark theme
)

# Show the plot
fig_indicator.show()
```

```python
# Import necessary libraries
import pandas as pd
import plotly.graph_objects as go
import json

# Load and read the posts JSON data
with open('data/instagram_analytics_posts.json') as f:
    posts_data = json.load(f)

# Convert to pandas DataFrame
posts_df = pd.DataFrame(posts_data)

# Calculate the average likes and comments per post
average_likes = posts_df['like_count'].mean()
average_comments = posts_df['comment_count'].mean()
total_posts = len(posts_df)  # Total number of posts

# Create a plotly figure for the bar charts
fig_bars = go.Figure()

# Add a bar for average likes
fig_bars.add_trace(go.Bar(
    x=['Average Likes per Post'],
    y=[average_likes],
    name='Average Likes',
    marker_color='deepskyblue',  # Change color if needed
    hovertemplate='Average Likes: %{y}<extra></extra>'  # Custom hover info
))

# Add a bar for average comments
fig_bars.add_trace(go.Bar(
    x=['Average Comments per Post'],
    y=[average_comments],
    name='Average Comments',
    marker_color='Red',  # Change color if needed
    hovertemplate='Average Comments: %{y}<extra></extra>'  # Custom hover info
))

# Customize layout for the bar chart
fig_bars.update_layout(
    title=f'Average Likes and Comments per Post (Total Posts: {total_posts})',
    title_x=0.5,
    hovermode='x',
    template='plotly_dark',  # Apply dark theme
    bargap=0.5  # Gap between bars
)

# Show the plot
fig_bars.show()
```

# Top 5 Posts with the Most Likes
- Description: Identify the publications that received the most likes.
- Visualization: Horizontal bar chart ordered by number of likes.



```python
# Import necessary libraries
import pandas as pd
import plotly.graph_objects as go
import json

# Load and read the posts JSON data
with open('data/instagram_analytics_posts.json') as f:
    posts_data = json.load(f)

# Convert to pandas DataFrame
posts_df = pd.DataFrame(posts_data)

# Ensure 'taken_at' is converted to datetime format
posts_df['taken_at'] = pd.to_datetime(posts_df['taken_at'], errors='coerce')  # Coerce any invalid dates to NaT

# Sort the posts by number of likes and select the top 5
top_5_posts = posts_df.nlargest(5, 'like_count')[['like_count', 'post_link', 'taken_at']]

# Create a plotly table for the top 5 posts with the most likes
fig_table = go.Figure(data=[go.Table(
    header=dict(values=['<b>Post Date</b>', '<b>Likes</b>', '<b>Post URL</b>'],
                fill_color='lightskyblue',
                align='left',
                font=dict(color='white', size=12)),
    cells=dict(values=[top_5_posts['taken_at'].dt.strftime('%Y-%m-%d'),  # Format the date properly
                       top_5_posts['like_count'],
                       top_5_posts['post_link']],
               fill_color='lightcyan',
               align='left',
               font=dict(color='black', size=11))
)])

# Customize layout for dashboard style
fig_table.update_layout(
    title='Top 5 Posts with the Most Likes',
    title_x=0.5,
    margin=dict(l=20, r=20, t=50, b=20),
    paper_bgcolor="rgba(0,0,0,0)",  # Transparent background
    plot_bgcolor="rgba(0,0,0,0)"    # Transparent background
)

# Show the table
fig_table.show()

# Display the top 5 posts in a table format (optional, for reference)
print(top_5_posts)
```

```python
# Import necessary libraries
import pandas as pd
import plotly.graph_objects as go
import json

# Load and read the posts JSON data
with open('data/instagram_analytics_posts.json') as f:
    posts_data = json.load(f)

# Convert to pandas DataFrame
posts_df = pd.DataFrame(posts_data)

# Ensure 'taken_at' is converted to datetime format
posts_df['taken_at'] = pd.to_datetime(posts_df['taken_at'], errors='coerce')  # Coerce any invalid dates to NaT

# Sort the posts by number of likes and select the top 5
top_5_posts = posts_df.nlargest(5, 'like_count')[['like_count', 'post_link', 'taken_at']]

# Define the layout for the cards
fig_cards = go.Figure()

# Add each post as a card using Annotations to simulate a card-like structure
for i, row in top_5_posts.iterrows():
    fig_cards.add_trace(go.Scatter(
        x=[0], y=[i],  # Positioning each card vertically
        text=(
            f"<b>Post Date:</b> {row['taken_at'].strftime('%Y-%m-%d')}<br>"
            f"<b>Likes:</b> {row['like_count']}<br>"
            f"<b>Post URL:</b> <a href='{row['post_link']}' target='_blank'>Click here</a>"
        ),
        mode='text',
        hoverinfo='skip',
        showlegend=False
    ))

# Customize the layout to simulate a card grid
fig_cards.update_layout(
    title='Top 5 Posts with the Most Likes',
    title_x=0.5,
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),  # Hide x-axis
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),  # Hide y-axis
    plot_bgcolor='rgba(0,0,0,0)',  # Transparent background
    paper_bgcolor='rgba(0,0,0,0)',  # Transparent background
    height=600,
    margin=dict(l=20, r=20, t=50, b=20),  # Adjust margins for dashboard layout
    annotations=[
        dict(
            x=0, y=i, text=f"<b>Post {i+1}</b>",
            showarrow=False, font=dict(size=14, color='black'),
            xref="paper", yref="y"
        ) for i in range(5)
    ]
)

# Show the "cards"
fig_cards.show()

# Display the top 5 posts in a table format (optional, for reference)
print(top_5_posts)
```

# Posts per Month
- Description: Show how many publications were made per month.
- Display: Stacked bar chart with months and post count.



```python
# Import necessary libraries
import pandas as pd
import plotly.express as px
import json

# Load and read the posts JSON data
with open('data/instagram_analytics_posts.json') as f:
    posts_data = json.load(f)

# Convert to pandas DataFrame
posts_df = pd.DataFrame(posts_data)

# Ensure 'taken_at' is converted to datetime format
posts_df['taken_at'] = pd.to_datetime(posts_df['taken_at'], errors='coerce')

# Extract the year and month from the 'taken_at' column for grouping
posts_df['year_month'] = posts_df['taken_at'].dt.to_period('M')

# Group the posts by year and month, and count the number of posts for each month
monthly_post_counts = posts_df.groupby('year_month').size().reset_index(name='post_count')

# Convert the 'year_month' back to datetime format for better visualization
monthly_post_counts['year_month'] = monthly_post_counts['year_month'].dt.to_timestamp()

# Create a bar chart to show the number of posts per month
fig_bar = px.bar(monthly_post_counts, 
                 x='year_month', 
                 y='post_count',
                 title='Number of Posts per Month',
                 labels={'year_month': 'Month', 'post_count': 'Number of Posts'},
                 template='plotly_dark')  # Use dark theme for dashboard

# Customize the bar chart layout
fig_bar.update_layout(
    xaxis_title='Month',
    yaxis_title='Number of Posts',
    title_x=0.5,
    bargap=0.2,  # Adjust gap between bars
    hovermode='x'
)

# Show the bar chart
fig_bar.show()
```

**Likes and Comments Growth Rate** Description: Show how the number of posts and comments has grown.
- Description: See how the number of interactions has grown over time.
- Visualization: Line graph showing the growth of likes and comments day by day.


```python
# Import necessary libraries
import pandas as pd
import plotly.express as px
import json

# Load and read the posts JSON data
with open('data/instagram_analytics_posts.json') as f:
    posts_data = json.load(f)

# Convert to pandas DataFrame
posts_df = pd.DataFrame(posts_data)

# Ensure 'taken_at' is converted to datetime format for time series analysis
posts_df['taken_at'] = pd.to_datetime(posts_df['taken_at'], errors='coerce')

# Sort the data by the date of the post (oldest to newest)
posts_df = posts_df.sort_values(by='taken_at', ascending=True)

# Create cumulative sums of likes and comments to show the growth over time
posts_df['cumulative_likes'] = posts_df['like_count'].cumsum()
posts_df['cumulative_comments'] = posts_df['comment_count'].cumsum()

# Create a new DataFrame for the line graph (group by day for clearer visualization)
daily_growth_df = posts_df.groupby(posts_df['taken_at'].dt.date).agg({
    'cumulative_likes': 'max', 
    'cumulative_comments': 'max'
}).reset_index()

# Create an interactive line graph using Plotly Express
fig_growth_rate = px.line(
    daily_growth_df,
    x='taken_at',  # X-axis: Date of the posts
    y=['cumulative_likes', 'cumulative_comments'],  # Y-axis: Cumulative likes and comments
    labels={
        'taken_at': 'Date',
        'value': 'Total Interactions',
        'variable': 'Interaction Type'
    },
    title='Interactions Growth Rate Over Time',
    template='plotly_dark'  # Dark theme
)

# Customize the appearance of the lines
fig_growth_rate.update_traces(mode='lines+markers')  # Add markers to the line graph

# Add hover information and interaction for better insights
fig_growth_rate.update_layout(
    hovermode='x unified',  # Show all hover information at once when hovering over a point
    xaxis_title='Date',
    yaxis_title='Cumulative Interactions',
    legend_title='Interaction Type'
)

# Show the interactive line graph
fig_growth_rate.show()
```

 **Life Span of Posts** Description: See how many days a post has been active over time.
- Description: See how many days a post continues to receive significant likes and comments.

```python
# Import necessary libraries
import pandas as pd
import plotly.express as px
import json

# Load and read the posts JSON data
with open('data/instagram_analytics_posts.json') as f:
    posts_data = json.load(f)

# Convert to pandas DataFrame
posts_df = pd.DataFrame(posts_data)

# Ensure 'taken_at' is converted to datetime format
posts_df['taken_at'] = pd.to_datetime(posts_df['taken_at'], errors='coerce')

# Sort by 'taken_at' to ensure posts are in chronological order
posts_df = posts_df.sort_values(by='taken_at', ascending=True)

# Set a threshold for significant interactions (e.g., 100 likes or comments)
significant_threshold = 100

# Define a function to calculate the lifespan of a post based on significant interactions
def calculate_lifespan(row):
    # Filter rows where the post continues to get significant interactions (likes or comments above threshold)
    interactions_df = posts_df[(posts_df['id'] == row['id']) & 
                               ((posts_df['like_count'] >= significant_threshold) | 
                                (posts_df['comment_count'] >= significant_threshold))]
    
    # If the post received significant interactions, calculate the days it remained active
    if not interactions_df.empty:
        first_day = row['taken_at']
        last_interaction_day = interactions_df['taken_at'].max()
        lifespan_days = (last_interaction_day - first_day).days
        return lifespan_days
    else:
        return 0  # If no significant interactions, life span is 0

# Apply the lifespan calculation to each post
posts_df['life_span'] = posts_df.apply(calculate_lifespan, axis=1)

# Filter out posts with 0 life span if desired
posts_df_filtered = posts_df[posts_df['life_span'] > 0]

# Create an interactive bar chart to visualize the life span of each post
fig_lifespan = px.bar(
    posts_df_filtered,
    x='taken_at',  # X-axis: Date of the posts
    y='life_span',  # Y-axis: Life span (days)
    hover_data=['post_link', 'life_span'],  # Hover data: Post URL and life span in days
    title='Life Span of Posts Based on Significant Interactions',
    labels={'taken_at': 'Post Date', 'life_span': 'Life Span (Days)'},
    template='plotly_dark'  # Dark theme
)

# Customize the layout for better visualization
fig_lifespan.update_layout(
    xaxis_title='Post Date',
    yaxis_title='Life Span (Days)',
    hovermode='x unified',  # Show all hover information at once when hovering over a point
)

# Show the interactive bar chart
fig_lifespan.show()
```
